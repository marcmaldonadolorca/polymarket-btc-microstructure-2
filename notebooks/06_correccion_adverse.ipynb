# Corrección del *score* de selección adversa

Este cuaderno documenta un **error que cometimos y corregimos**. Lo dejamos a propósito
en el repositorio, porque la lección metodológica vale más que el resultado.

## Qué pasó

En los experimentos de selección usábamos una variable objetivo de *selección adversa*:
una etiqueta en la que el valor **positivo señala un resultado desfavorable** (una
ejecución que sale mal). Al construir la puntuación que ordena las operaciones, la
usamos tal cual, de modo que la probabilidad de adverso actuaba **al revés** de como
debía: puntuábamos mejor justo lo que había que evitar.

## La corrección

La orientación correcta del score es

```text
safe_adverse_score = 1 - P(adverso)
```

es decir, una operación es buena cuando la probabilidad de resultado adverso es **baja**.
Tras invertir el signo, marcamos como superados todos los resultados afectados y
repetimos la validación. La candidata que teníamos hasta entonces **dejó de valer como
evidencia**.

## Por qué lo contamos

Porque una métrica positiva no basta. Es fácil obtener un número bonito con una
puntuación que, mirada de cerca, está orientada al revés. La conclusión que nos llevamos
—y que aplicamos al resto del proyecto— es que hay que **auditar el significado
económico de la señal que guía la selección**, no solo su valor. Ver
`docs/COMPLEX_V1A_CLEAN_SELECTION_PROBE.md`.
