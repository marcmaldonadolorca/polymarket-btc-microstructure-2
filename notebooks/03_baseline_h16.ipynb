## Resultados cargados

Las celdas siguientes leen los CSV que generó el script, para que el cuaderno se pueda
navegar sin tener que mirar todo el log de ejecución:

- `final_summary.csv` — el resumen por familia de features (¿qué bloque aporta?);
- `walkforward_summary.csv` — estabilidad a lo largo de los folds temporales;
- `rule_summary.csv` — el resultado de la regla de decisión;
- `best_model_segments.csv` — desglose por segmentos del mejor modelo;
- `best_model_permutation_importance.csv` — qué features pesan de verdad
  (importancia por permutación, interpretabilidad ligera).


```python
from pathlib import Path
import json
import pandas as pd
from IPython.display import display

RESULTS = Path('data/experiments/baseline_v0_full_core_robustness/results')
summary = json.loads((RESULTS / 'summary.json').read_text(encoding='utf-8'))
print(json.dumps(summary, indent=2)[:4000])
```

```python
final_summary = pd.read_csv(RESULTS / 'final_summary.csv')
cols = ['model_kind','feature_set','selected_threshold','test_balanced_accuracy','test_macro_f1','test_action_actions','test_action_coverage_pct','test_action_hit_up_pct','test_action_wrong_down_pct','test_action_net_visible_mean','test_action_net_positive_pct','fit_seconds']
display(final_summary[cols].round(4))
```

```python
walk = pd.read_csv(RESULTS / 'walkforward_summary.csv')
cols = ['fold','model_kind','balanced_accuracy','macro_f1','selected_threshold','actions','coverage_pct','hit_up_pct','wrong_down_pct','net_visible_mean','net_positive_pct','fit_seconds']
display(walk[cols].round(4))
```

```python
rule = pd.read_csv(RESULTS / 'rule_summary.csv')
cols = ['label','split','complete_frames','actions','coverage_pct','hit_up_pct','flat_pct','wrong_down_pct','delta_mean','cost_median','net_visible_mean','net_positive_pct']
display(rule[cols].round(4))
```

```python
segments_path = RESULTS / 'best_model_segments.csv'
if segments_path.exists():
    seg = pd.read_csv(segments_path)
    display(seg.round(4))
```

```python
perm_path = RESULTS / 'best_model_permutation_importance.csv'
coef_path = RESULTS / 'logreg_top_coefficients.csv'
if perm_path.exists():
    display(pd.read_csv(perm_path).head(30).round(6))
if coef_path.exists():
    display(pd.read_csv(coef_path).head(30).round(6))
```
